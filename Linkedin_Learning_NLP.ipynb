## Regular Expressions


**'[j-q]' **-->> searches between j & q. returns one character at a time 

***example :*** if we wanted to search "nlp" in '[j-q]' regular expression, it returns 

n

l

p






**'[j-q]+'** --> returns all the charactes in a sequence or returns string

example : when searching "nlp", it returns -->>   n l p 

**'[j-q 0-9]+'** --> when seaching "nlp 2017" , it returns --->>  nlp2017

```
import re

re_test = "This is made up string to test 2different regex methods"
re_test
```

####Step 1- Removing Punctuation (by importing string)

```
#function to remove punctuations
def remove_punct(text):
  #joing the charracters by whitespaces
  text_nopunct = "".join([char for char in text if char not in string.punctuation])
  return text_nopunct

#creating new column "body_text_clean"
data["body_text_clean"] = data["body_text"].apply(lambda x: remove_punct(x))

#displaying 1st 5 rows
data.head()
```

####Step 2 - **Tokenization** (by importing re)

+ splitting sentences into words of list

```
import re

def tokenize(text):
  tokens = re.split("\W+", text) #splitting by non words
  return tokens

data["body_text_tokenized"] = data["body_text_clean"].apply(lambda x : tokenize(x.lower())) #converting the text to lower case
                                                                                            #and then appply tokenize function
data.head()
```

Step 3 -  **Removing Stopwords** (by importing nltk package)

```
import nltk
nltk.download('stopwords')

stopword_list = nltk.corpus.stopwords.words("english")
stopword_list
```

```
def remove_stopwords(tokenized_text):
  text = [word for word in tokenized_text if word not in stopword_list]
  return text

data["body_text_nostopwords"] = data["body_text_tokenized"].apply(lambda x: remove_stopwords(x))
data.head()
```

# Count Vectorization

creates the document-term matrix where each cell contains the count of number of times words occured(repeated) in that document

###Reading Raw text

```
#reading the file

import pandas as pd

pd.set_option("display.max_colwidth", 100) #setting column width = 100, by default its 50

data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/SMSSpamCollection.tsv", sep="\t" ,header= None)
data.columns = [ "label", "body_text"]

data.head()
```

## Clean up text

```
# Porter Stemmer
import nltk
nltk.download('stopwords')
ps = nltk.PorterStemmer()

import re
import string

stopword_list = nltk.corpus.stopwords.words("english")
stopword_list

def clean_text(text):
  #removing punctutaions
  text = "".join([word for word in text if word not in string.punctuation])
  #tokenization
  tokens = re.split("\W+", text)
  #removing stopwords and stemming
  text = [ps.stem(word) for word in tokens if word not in stopword_list]
  return text 
```

## Apply CountVectorizer

```
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer=clean_text) #calling function 
X_counts = count_vect.fit_transform(data["body_text"])

print(X_counts)
print("*****************")
print(X_counts.shape)
print("*****************")
print(count_vect.get_feature_names())
print("*****************")
print(count_vect)
```

N-Grams

```
#reading the file

import pandas as pd

pd.set_option("display.max_colwidth", 100) #setting column width = 100, by default its 50

data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/SMSSpamCollection.tsv", sep="\t" ,header= None)
data.columns = [ "label", "body_text"]

data.head()
```

Clean up text

```

# Porter Stemmer
import nltk
nltk.download('stopwords')
ps = nltk.PorterStemmer()

import re
import string

stopword_list = nltk.corpus.stopwords.words("english")
stopword_list

def clean_text(text):
  #removing punctutaions
  text = "".join([word for word in text if word not in string.punctuation])
  #tokenization
  tokens = re.split("\W+", text)
  #removing stopwords and stemming
  text = " ".join([ps.stem(word) for word in tokens if word not in stopword_list])
  return text 
```

```
data["cleaned_text"] = data["body_text"].apply(lambda x : clean_text(x.lower()))

data.head()
```

Applying countvectorizer (N - Grams)

```
from sklearn.feature_extraction.text import CountVectorizer

ngram_vector = CountVectorizer(ngram_range=(2,2)) #bi gram
X_counts = ngram_vector.fit_transform(data["cleaned_text"])

print(X_counts.shape)
print("************")
print(ngram_vector.get_feature_names())
```

```
X_counts_df = pd.DataFrame(X_counts.toarray())
X_counts_df.columns = ngram_vector.get_feature_names()
print(X_counts_df) # 1 mesna it word appeared in the text
```
